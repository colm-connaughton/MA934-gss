## Golden section search

Given a function, $f(x)$ of a single variable, the task is to find a minimum of $f$. 

An ordered triple of points $(a,c,b)$ is said to bracket a minimum of $f$ if $f(x) < f(a)$ and $f(x) < f(c)$.

```julia
push!(LOAD_PATH, pwd())
using LaTeXStrings
using PyPlot
using Revise
using MA934Figures
```

Given a bracketing triple, $(a, c, b)$, we evaluate $f$ at a new point, $x$, and use this value to construct a new, smaller bracketing triple. We then iterate this process until a desired degree of accuracy is reached. At each step, it is clearly most efficient to choose $x$ in the larger of the two subintervals defined by the triple $(a,c,b)$. This gives two cases, each of which yields two possibilities for the new bracketing triple as illustrated in this figure:

```julia
fig, (ax1, ax2) = gssfigure1();
```

If $[a, c]$ is the larger subinterval (Case 1) then the new new bracketing triple is either $(a,x,c)$ (Case 1(A)) or $(x,c,b)$ (Case 1(B)). We want the width of the new bracketing triple to be independent of which of these outcomes occurs. Thus we choose:

>$c - a = b - x$

which gives

>$x = a + b - c$.

Similarly, if $[c,b]$ is the larger subinterval (Case 2), we want the width of the new bracketing triple to be independent of whether Case 2(A) or Case 2(B) occurs:

>$x - a = b - c$

which again gives

>$x = a + b -c$.

Thus regardless of whether we are in Case 1 or Case 2, we always choose $x = a + b - c$.


The second part of the argument is to decide where to place $c$ in the initial bracketing triple. Let $p$ denote the width of the longer of the two subintervals of the initial bracketing triple, and let $q$ denote the width of the shorter subinterval. This is as shown above. Let $r$ denote the shorter subinterval in the new bracketing triple.

The idea of the golden section search is to choose $c$ such that the ratio of the width of the shorter subinterval to the width of the longer one always remains constant as we go from the initial bracketing triple to the new one.

If Case 1 occurs, this means we must have:

> Case 1(A) : $\frac{r}{p-r} = \frac{q}{p}$

> Case 1(B) : $\frac{r}{q} = \frac{q}{p}$

If Case 2 occurs, we arrive at the same equations:

> Case 2(A) : $\frac{r}{q} = \frac{q}{p}$

> Case 2(B) : $\frac{r}{p-r} = \frac{q}{p}$

Thus regardless of which case we are in, eliminating $r$ from these equations gives

> $\left(\frac{q}{p} \right)^2 + \left(\frac{q}{p} \right) - 1 = 0$.

The solution is

> $\left(\frac{q}{p} \right) = \frac{\sqrt{5}\pm 1}{2}$.

We choose the negative sign because we had assumed that $q$ was width of the shorter subinterval so $q/p$ should be less than one. Thus we choose $c$ such that the ratio $q/p$ is given by the golden ratio. Remember that $x$ has already been determined above: provided we start with the correct ratio, $q/p$, this ratio will be preserved when we construct the new triple from evaluating f at $x = a + b - -c$.



An easy way to avoid multiple case checks when implementing this is to use temporary variables $x_1$ and $x_2$ to store the points $x$ and $c$ (in Case 1) or $c$ and $x$ (in Case 2). We also use temporary variables $f_1$ and $f_2$ to store the associated function values. This then combines cases 1 and 2 into a single arrangement so that there are only 2 possible cases. Regardless of the order in which previous points were evaluated, the new bracketing triple will consist of the three points centred on the point with the smallest value of $f$ found so far. Thus by comparing $f_1$ to $f_2$, there are two cases:

```julia
fig, ax1 = gssfigure2();
```

Let us starting from a bracketing triple $(a,c,b)$ and with function values $(f_a, f_c, f_b)$. We assume that this triple already has the ratio $q/p = \phi$. The pseudocode to construct the new bracketing triple goes as follows:

x = a + b - c  
$f_x = f(x)$  

if c - a > b - c
> $x_1 \leftarrow x$  
> $x_2 \leftarrow c$  
> $f_1 \leftarrow f_x$  
> $f_2 \leftarrow f_c$

else

> $x_1 \leftarrow c$  
> $x_2 \leftarrow x$  
> $f_1 \leftarrow f_c$  
> $f_2 \leftarrow f_x$

end

if $f_1$ < $f_2$

> (a,c,b) $\leftarrow$ (a, $x_1$, $x_2$)  
> ($f_a$, $f_c$, $f_b$) $\leftarrow$ ($f_a$, $f_1$, $f_2$)  

else

> (a,c,b) $\leftarrow$ ($x_1$, $x_2$, b)  
> ($f_a$, $f_c$, $f_b$) $\leftarrow$ ($f_1$, $f_2$, $f_b$)

end

Here is some Julia code that implements this. It tracks both the position of estimate of the miniumum and the width of the bracketing triple for plotting later.

```julia
function gss(f, triple, tol)
    # Some arrays to store intermediate values for plotting
    steps = Int64[]
    position = Float64[]
    width = Float64[]
    
    # Initialise
    (a,b,c) = triple
    (fa, fc, fb) = (f(a), f(c), f(c))
    
    n = 0 
    
    while(b-a > tol && n <30)
        x = a + b -c
        fx = f(x)
        
        append!(steps,n)
        append!(position, x)
        append!(width, b-a)
        
        if (c-a < b-c)
            x1 = c; f1 = fc
            x2 = x; f2 = fx
        else
            x1=x; f1 = fx
            x2=c; f2 = fc
        end
        
        if (f1 < f2)
            c = x1
            b = x2
            fc = f1
            fb = f2
        else
            a = x1
            c = x2
            fa = f1
            fc = f2
        end
        n+=1
        
    end
    return steps, position, width
end
```

Let us now use this to find the minimum of $f(x) = x^2$ at $x=0$

```julia
# Define the function to be minimised
function f(x)
    return x^2
end


# Construct initial bracketing triple
ϕ = 0.5*(sqrt(5.0)-1.0)
a = -1.0
b = 1.0
c = b - (b-a)*ϕ
t0 = (a,b,c)

# Run the golden section search
steps, position, width = gss(f, t0, 1.0E-5);
```

The following plots illustrate that the algorithm is behaving as expected:

```julia
gssfigure3(steps, position, width)
```
